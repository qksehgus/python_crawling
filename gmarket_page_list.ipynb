```python
from bs4 import BeautifulSoup
import requests

def url_input():
    input_data=input("원하시는 제품을 검색 하시오 : ")
    url='http://browse.gmarket.co.kr/search?keyword='+input_data
    return url

def url_encoding(url):
    encode = requests.get(url =url)
    base = BeautifulSoup(encode.content,"html.parser")
    return base

def re_title(http):
    title=list()
    x=http.findAll("span",{"class":"text__item"})
    for i in x:
        title.append(i.text)
    return title

x=url_input()
http=url_encoding(x)
all_Title=list()
all_Title.append(re_title(http))

print(all_Title)
```
