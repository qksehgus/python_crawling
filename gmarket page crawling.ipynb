```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import json

def url_input(input_page):
    input_data=input("원하시는 제품을 검색 하시오 : ")
    re_list=list()
    for i in range(1,int(input_page)):
        url='http://browse.gmarket.co.kr/search?keyword='+input_data+'&k=32&p='+str(i)
        re_list.append(url)
        
    return re_list



def title_def(url):
    title=list()
    encode = requests.get(url =url)
    base = BeautifulSoup(encode.content,"html.parser")
    x=base.findAll("span",{"class":"text__item"})
    for i in x:
        title.append(i.text)
    
    return title

def price_def(url):
    title=list()
    encode = requests.get(url =url)
    base = BeautifulSoup(encode.content,"html.parser")
    x=base.findAll("strong",{"class":"text text__value"})
    for i in x:
        title.append(i.text)
    
    return title
```

```python
input_page=input("원하는 페이지 수를 입력하시오 : ")
x=url_input(input_page)
title_list=list()
price_list=list()
product_list=list()
title_list.append(title_def(x[1]))
price_list.append(price_def(x[1]))     


for i in range(0,int(input_page)-1):
    title_list.append(title_def(x[i]))
    price_list.append(price_def(x[i]))
    
for i in range(0,int(input_page)):
    for l in range(0,100):
        product=([title_list[i][l],price_list[i][l]])
        product_list.append(product)
```
